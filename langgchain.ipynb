# Text Summarization

```python
# i want to perform text summarisation
from langchain_core.prompts import PromptTemplate

template = """You are an expert disaster report summarizer.  
Summarize the given text about a natural calamity into a clear and concise paragraph,Don't add any extra information.  

The summary must include, if available:
- Date, day or month of the calamity
- Location (village, district, state, country, etc.)
- Duration (how long it lasted)
- Number of lives lost
- Infrastructure damaged (roads, bridges, houses, power, schools, etc.)
- Losses for farmers (crops, livestock, farmland, money impact)
- Total estimated loss  

Write the output as **one well-structured paragraph**.

### Example Input:
"A massive flood hit Bihar on August 10, lasting for 5 days. Over 200 people died, and thousands of houses and bridges were destroyed. Farmers reported crop damage worth 500 crore rupees. The government announced 10,000 rupees compensation per affected family and set up relief camps."

### Example Output:
On August 10, a devastating flood struck Bihar and lasted for five days, claiming over 200 lives. The disaster caused widespread destruction, washing away thousands of houses and bridges, while farmers suffered agricultural losses worth 500 crore rupees. In response, the government provided relief measures, including 10,000 rupees in compensation for each affected family and the setup of multiple relief camps.

---

Now, summarize the following text:
{input_text}
"""
prompt = PromptTemplate(template = template,
               input_variable = ["input_text"],
               )
```

```python
# data for text summarization
import pandas as pd
text_data = pd.read_excel("C:/Users/Admin/Desktop/Kiran_DS_Lab/Agentic_ai/Duplicate_Using_consine_Matrix_LLM.xlsx")
chain = prompt|model|output_parser
text_data["Summary"] = text_data["text"].apply(lambda x: chain.invoke(x))
```
