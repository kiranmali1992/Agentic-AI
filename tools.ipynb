```python
# developing custom tools
# create tool to reserse string
from langchain.tools import tool
@tool
def reserve_string(input_string:str) ->str:
    """this function is reversing string"""
    return input_string[::-1] 
```

```python
reserve_string.invoke("kiran")
```
